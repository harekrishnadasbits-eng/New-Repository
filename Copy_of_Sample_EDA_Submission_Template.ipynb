An ECDF (Empirical Cumulative Distribution Function) plot is chosen to visualize the cumulative distribution of 'Line Item Value'. ##### 2. What is/are the insight(s) found from the chart?

For example, we can see that a large proportion (e.g., 80-90%) of transactions have values below a certain threshold. The curve rises steeply at lower values and then flattens out, indicating that higher 'Line Item Values' are much less frequent and contribute to the long tail of the distribution, confirming the skewed nature of the data.

##### 3. Will the gained insights help creating a positive business impact?
Yes, the insights from the ECDF plot are very valuable for business impact. By precisely knowing the cumulative distribution of 'Line Item Value', businesses can: 1) Understand the typical scale of transactions and design strategies (e.g., promotional offers, processing workflows) that cater to the majority of transactions. 2) Set realistic targets and benchmarks by understanding various percentiles (e.g., what is the 90th percentile 'Line Item Value'). 3) Identify thresholds for special handling – for instance, transactions above the 95th percentile might warrant senior approval or additional scrutiny.
````python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
````

##### 1. Why did you pick the specific chart?
A correlation heatmap is chosen to visualize the relationships between all pairs of numerical variables in the dataset. It provides a quick and intuitive way to identify strong positive, negative, or no correlations, which is crucial for understanding multivariate relationships and potential multicollinearity.

##### 2. What is/are the insight(s) found from the chart?
The heatmap reveals the strength and direction of linear relationships between numerical features. For example, 'Line Item Quantity' and 'Line Item Value' might show a strong positive correlation, indicating that larger quantities generally lead to higher values. 'Pack Price' and 'Unit Price' should also be highly correlated. We can identify variables that are highly correlated with the target variable (if any), as well as highly correlated independent variables (which might indicate multicollinearity issues for modeling).

##### 3. Will the gained insights help creating a positive business impact?
Yes, the insights from a correlation heatmap are highly valuable for business impact. By understanding these relationships, businesses can: 1) Optimize pricing strategies: If 'Pack Price' and 'Unit Price' are strongly correlated, it confirms consistent pricing logic. 2) Improve forecasting: Strongly correlated variables can be used in predictive models. 3) Resource allocation: If 'Line Item Quantity' drives 'Line Item Value', efforts could focus on increasing order sizes. 4) Data cleaning and feature engineering: Identify redundant features (highly correlated independent variables) or create new features by combining related ones. Is there a difference in delivery performance (on-time vs. delayed) based on the time between the PO Sent to Vendor Date and the Scheduled Delivery Date?

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure date columns are in datetime format, handling potential errors
df_clean['PO Sent to Vendor Date'] = pd.to_datetime(df_clean['PO Sent to Vendor Date'], errors='coerce')
df_clean['Scheduled Delivery Date'] = pd.to_datetime(df_clean['Scheduled Delivery Date'], errors='coerce')
df_clean['Delivered to Client Date'] = pd.to_datetime(df_clean['Delivered to Client Date'], errors='coerce')

# Recalculate 'Delivery On Time' to ensure consistency with current df_clean state
df_clean['Delivery On Time'] = df_clean['Delivered to Client Date'] <= df_clean['Scheduled Delivery Date']

# Calculate the lead time (in days) between PO Sent and Scheduled Delivery
df_clean['Lead Time Days'] = (df_clean['Scheduled Delivery Date'] - df_clean['PO Sent to Vendor Date']).dt.days

# Remove rows where Lead Time Days is negative or NaN (invalid dates)
df_clean_lead_time = df_clean.dropna(subset=['Lead Time Days']).copy()
df_clean_lead_time = df_clean_lead_time[df_clean_lead_time['Lead Time Days'] >= 0]

# Define bins for lead time categories
bins = [-1, 30, 90, 180, 365, df_clean_lead_time['Lead Time Days'].max()]
labels = ['<1 Month', '1-3 Months', '3-6 Months', '6-12 Months', '>12 Months']
df_clean_lead_time['Lead Time Category'] = pd.cut(df_clean_lead_time['Lead Time Days'], bins=bins, labels=labels, right=True)

# Group by Lead Time Category and calculate the percentage of on-time deliveries
on_time_performance_by_lead_time = df_clean_lead_time.groupby('Lead Time Category')['Delivery On Time'].mean().reset_index()
on_time_performance_by_lead_time['On-time Percentage'] = on_time_performance_by_lead_time['Delivery On Time'] * 100

# Sort categories logically if they are not already ordered by `pd.cut`
on_time_performance_by_lead_time['Lead Time Category'] = pd.Categorical(on_time_performance_by_lead_time['Lead Time Category'], categories=labels, ordered=True)
on_time_performance_by_lead_time = on_time_performance_by_lead_time.sort_values('Lead Time Category')

plt.figure(figsize=(12, 7))
sns.barplot(x='Lead Time Category', y='On-time Percentage', data=on_time_performance_by_lead_time, palette='plasma')
plt.title('On-time Delivery Percentage by Lead Time Category')
plt.xlabel('Lead Time Category (PO Sent to Scheduled Delivery)')
plt.ylabel('On-time Delivery Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?
A bar chart is suitable for comparing the on-time delivery percentages across different discrete categories of lead time. It allows for a clear visual comparison of how lead time impacts the punctuality of deliveries.

##### 2. What is/are the insight(s) found from the chart?
The chart will show whether shorter or longer lead times tend to result in higher or lower on-time delivery percentages. For instance, we might find that very short lead times lead to more delays, or conversely, that very long lead times introduce more variables and thus more delays. This analysis will highlight the optimal lead time windows for maintaining high delivery performance.

##### 3. Will the gained insights help creating a positive business impact?
Yes, these insights are crucial for optimizing planning and scheduling within FedEx Logistics. By understanding the relationship between lead time and on-time delivery, the business can: 1) **Optimize order placement:** Adjust procurement strategies to align with lead times that maximize on-time performance. 2) **Set realistic expectations:** Inform clients about realistic delivery timelines based on lead time categories. 3) **Identify bottlenecks:** If a particular lead time category shows significantly lower on-time performance, it can prompt further investigation into the causes (e.g., capacity constraints, vendor performance). 4) **Improve efficiency:** Focus on processes that are efficient for different lead time durations.
